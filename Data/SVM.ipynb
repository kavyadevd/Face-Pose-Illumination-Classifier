## PCA

```python
from pca import *
```

```python
# Try with different number of components
num_comp = [10,30,50,70,85]
train_x = data_all[0]
test_x = data_all[2]
acc_vec=[]
arrc_vec = []
# Resize to 2D:
#train_x = train_x.reshape(train_x.shape[0]*train_x.shape[1],train_x.shape[2])
for num in num_comp:
    pca = PCA(num)
    pca.fit(train_x)
    train_x_dash = pca.transform(train_x)
    # pca = PCA(num)
    # pca.fit(test_x)
    test_x_dash = pca.transform(test_x)
    # print(train_x_dash.shape)
    # print(test_x_dash.shape)
    train_x_dash = train_x_dash.reshape(train_x_dash.shape[0]//2,2,num)

    cross_validation_arr = [0.1,0.3,0.4, 0.6, 1.1, 1.6, 2.1, 2.6, 3.1, 3.6, 4.5,5,6]
    classifier = SVM('rbf', cross_validation_arr, 0.001, 6)
    train_x_dash = train_x_dash.reshape(train_x_dash.shape[0]*train_x_dash.shape[1],train_x_dash.shape[2])
    acc,arrc = classifier.classify(train_x_dash,data_all[1],test_x_dash,data_all[3],10)
    acc_vec.append(acc)
    arrc_vec.append(arrc)
```

```python
arrc_vec
```

```python

plt.xlabel('C')
plt.ylabel('Accuracy')
plt.ylim(ymin=0)
for i in range(len(num_comp)):
    leg = 'Num of comp :'+ str(num_comp[i])
    plt.plot(arrc_vec[i], acc_vec[i], label = leg)
plt.legend()
plt.show()
```

## MDA

```python
from mda import *
```

```python
train_x = data_all[0]
test_x = data_all[2]
acc_vec_mda_rbf = []
arrc_vec_mda_rbf = []

num_newdim = [10, 30, 85, 90, 150]
for n_dim in num_newdim:
    mda_data = MDA(0.6, train_x, test_x, n_dim)
    mda_data.get_matA()
    train_x_dash = mda_data.transform(train_x)
    test_x_dash = mda_data.transform(test_x)

    cross_validation_arr = [0.1, 0.3, 0.4, 0.6, 2.6, 6]
    classifier = SVM('rbf', cross_validation_arr, 0.001, 1)
    #train_x_dash = train_x_dash.reshape(train_x_dash.shape[0]*train_x_dash.shape[1],train_x_dash.shape[2])
    acc, arrc = classifier.classify(
        train_x_dash, data_all[1], test_x_dash, data_all[3], 10)
    acc_vec_mda_rbf.append(acc)
    arrc_vec_mda_rbf.append(arrc)
```

```python
plt.title('SVM RBF Kernel with MDA for Neutral-Exp classification')
plt.xlabel('C')
plt.ylabel('Accuracy')
#plt.ylim(ymin=0)
for i in range(len(num_newdim)):
    leg = 'Num of comp :'+ str(num_newdim[i])
    plt.plot(arrc_vec_mda_rbf[i], acc_vec_mda_rbf[i], label = leg)
plt.legend()
plt.show()
```
