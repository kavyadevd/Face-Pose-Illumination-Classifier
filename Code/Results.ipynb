```python
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import numpy as np
```

```python
acc_bayes_data_orig = 0.74 #333333333333333
acc_bayes_data_pca = []
acc_bayes_data_mda = []

acc_bayes_pose_orig = 0.87 #66666666666667
acc_bayes_pose_pca = []
acc_bayes_pose_mda = []

acc_bayes_illum_orig = 1
acc_bayes_illum_pca = []
acc_bayes_illum_mda = []

x_labels = ['none','PCA','MDA']
```

```python

acc_bayes_data_pca_best = acc_bayes_data_pca[np.argmax(acc_bayes_data_pca)]
acc_bayes_data_mda_best = acc_bayes_data_mda[np.argmax(acc_bayes_data_mda)]

y_acc = [acc_bayes_data_orig,acc_bayes_data_pca_best,acc_bayes_data_mda_best]
plt.title('Bayes classification for Face dataset')
plt.xlabel('Dim reduction')
plt.ylabel('Accuracy')
plt.bar(x_labels, y_acc, color ='maroon', width = 0.4)
plt.legend()
plt.show()
```

```python

acc_bayes_pose_pca_best = acc_bayes_pose_pca[np.argmax(acc_bayes_pose_pca)]
acc_bayes_pose_mda_best = acc_bayes_pose_mda[np.argmax(acc_bayes_pose_mda)]

y_acc = [acc_bayes_pose_orig,acc_bayes_pose_pca_best,acc_bayes_pose_mda_best]
plt.title('Bayes classification for Pose')
plt.xlabel('Dim reduction')
plt.ylabel('Accuracy')
plt.bar(x_labels, y_acc, color ='maroon', width = 0.4)
plt.legend()
plt.show()
```

```python

acc_bayes_illumination_pca_best = acc_bayes_illumination_pca[np.argmax(acc_bayes_illumination_pca)]
acc_bayes_illumination_mda_best = acc_bayes_illumination_mda[np.argmax(acc_bayes_illumination_mda)]

y_acc = [acc_bayes_illumination_orig,acc_bayes_illumination_pca_best,acc_bayes_illumination_mda_best]
plt.title('Bayes classification for illumination')
plt.xlabel('Dim reduction')
plt.ylabel('Accuracy')
plt.bar(x_labels, y_acc, color ='maroon', width = 0.4)
plt.legend()
plt.show()
```
