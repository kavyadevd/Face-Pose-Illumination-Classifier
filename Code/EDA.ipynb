```python
%matplotlib inline
from scipy.io import loadmat
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
mat = loadmat('data.mat')
data = mat['face']
```

#### Dataset Visualization

```python
# a = data[:,:,3*1-2]
# plt.imshow(a)
plt.figure(figsize=(50, 50))
samples_per_class = 3
for n in range(1,12):
    x1 = data[:,:,3*n-2]
    x2 = data[:,:,3*n-1]
    x3 = data[:,:,3*n]
    plt.subplot(12,3, n)
    plt.imshow(x1)
    plt.subplot(12,3, n+1)
    plt.imshow(x2)
    plt.subplot(12,3, n+2)
    plt.imshow(x3)
    plt.axis('off')
plt.show()
#plt.show()
```
